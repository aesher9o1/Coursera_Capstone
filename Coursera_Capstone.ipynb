```python
import pandas as pd
import urllib3
from bs4 import BeautifulSoup
```

```python
#all the constants goes here
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

Let us now create a pool manager for URLLib 3 and use it to store the page in a variable called soup. It must be noted that simply requesting the URL will give **InsecureRequestWarning: Unverified HTTPS request is being made. Adding certificate verification is strongly advised.** Hence we will add a ignore security warning

```python
http = urllib3.PoolManager()
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
response = http.request('GET', url)
soup = BeautifulSoup(response.data).find('table')
table_rows = soup.find_all('tr')
table_rows
```

```python
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
    
l    
```

Now let us make a dataframe

```python
df = pd.DataFrame(l, columns=["Postcode", "Borough", "Neighbourhood"])
df.drop(df.index[[0]], inplace=True)
df
```

```python
Next up we will drop all the not assigned Borough. Reset the index and replace the \n that was coming
```

```python
df = df[~df.Borough.str.contains("Not assigned")]
df= df.reset_index()
df = df.iloc[:,1:]
df["Neighbourhood"]= df["Neighbourhood"].str.replace("\n", "") 
df
```

```python
for index, row in df.iterrows():
    if(row['Neighbourhood']=="Not Assigned"):
        row['Neighbourhood'] = row['Borough']

df
```

```python
df = df.groupby(['Postcode'])['Neighbourhood'].apply(', '.join).reset_index()
df
```
